```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "traj_nve.xyz"
atoms = []
coordinates = []
xyz = open(filename)
for frame in range(801):
    f = []
    n_atoms = int(xyz.readline())
    title = xyz.readline()
    for i in range(216):
        mol = []
        for j in range(5):
            line = xyz.readline()
            #print(line)
            atom,x,y,z = line.split()
            mol.append( [float(x), float(y), float(z), atom])
        f.append(mol)
    coordinates.append(f)
xyz.close()

print("filename:         %s" % filename)
print("title:            %s" % title)
print("number of atoms:  %d" % n_atoms)
```

```python
def dis(cord1, cord2):
    res = 0
    for i in range(3):
        d = abs(cord1[i]-cord2[i])
        if d > 11.9:
            d -= 11.9
        res += d**2
    return np.sqrt(res)
```

```python
dum = [0]*119
for frame in range(801):
    for i in range(0, 215):
        for j in range(j, 216):
            distance = dis(coordinates[frame][i][0], coordinates[frame][j][0])
            bin = int(distance*10)
            if bin<119:
                dum[bin] += 1

f = open("data.dat",'w')
g = [0]*119
for i in range(1,119):
    r = i*0.1
    g[i] = dum[i]/(2*0.016*r*0.1*np.pi*111.5*216)
    f.write(str(r) + " " + str(g[i]) + "\n")
f.close()
```
